```python
# Importando biblioteca pandas no Python
import pandas as pd

combustiveis_df = pd.read_excel("ca202102_20230207120945.xlsx")
```

```python
# Usa o print para ver o dataframe
print(combustiveis_df)
```

```python
display(combustiveis_df)
```

```python
# Head exibe as primeiras 5 linhas 
display(combustiveis_df.head())
```

```python
# Quero na verdade exibir as primeiras 15 linhas
display(combustiveis_df.head(15))
```

```python
# Comandos Dataframe.shape, Dataframe.info() e Dataframe.describe()

# Shape mostra o número de linhas e colunas (linhas, colunas)
print(combustiveis_df.shape)


# Shape[0] mostra só as linhas
print(combustiveis_df.shape[0])
```

```python
# Info() mostra quais são as colunas e que tipo de dados cada um tem... 
print(combustiveis_df.info())
```

```python
# Describe() mostra as estatisticas "mais basicas" que agente precisa
print(combustiveis_df.describe())
```

```python
# Filtrar apenas por uma coluna
display(combustiveis_df['Revenda'])
```

```python
# Aqui criamos um novo dataframe apenas com as colunas que eu quero
ca_df = combustiveis_df[['Revenda', 'Municipio', 'Produto', 'Valor de Venda']]
display(ca_df)
```

```python
# Loc[] exibe a 4a. linha
display(ca_df.loc[3])
```

```python
# Exibir a 10a. linha ate a 20a. linha
display(ca_df.loc[9:19])
```

```python
# Criar uma dataframe gas_df contendo apenas as 4 colunas (Revenda, Municipio, Produto, Valor de venda),
# somente com combustivel sendo GASOLINA e exbir na tela.
gas_df = ca_df.loc[ca_df['Produto'] == 'GASOLINA']
display(gas_df)

display(gas_df['Valor de Venda'].max())
```

```python
# Se eu quiser saber qual é o posto e muicipio da gasolina mais cara do pais?
display(gas_df[['Revenda', 'Municipio', 'Valor de Venda']].max())
```

```python
####### PERGUNTAR AO PROFESSOR 


# Dataframe.loc[] com multiplas condições para filtragem
# Quais são os preços, postos que vendem ETANOL na minha cidade (INDAIATUPA)
etanol_indaiatuba_df = ca_df.loc[(ca_df['Produto'] == 'ETANOL') & (ca_df['Municipio'] == 'INDAIATUPA')]
display(etanol_indaiatuba_df.sort_values(by='Valor de Venda'))
```

```python
# Dataframe.loc[] com multiplas condições para filtragem
# Quais são os preços, postos que vendem ETANOL na minha cidade (INDAIATUPA)
etanol_indaiatuba_df = ca_df.loc[(ca_df['Produto'] == 'ETANOL') & (ca_df['Municipio'] == 'INDAIATUBA')]
display(etanol_indaiatuba_df.sort_values(by='Valor de Venda'))
```

```python
from numpy import mean
# Qual a media de preços dos combustiveis GASOLINA e GASOLINA ADTIVADA do Bairro MOOCA em SÃO PAULO?

display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') & 
                            (combustiveis_df['Municipio'] == 'SAO PAULO') & 
                            ((combustiveis_df['Produto'] == 'GASOLINA') | (combustiveis_df['Produto'] == 'GASOLINA ADITIVADA')),
                            ['Valor de Venda']].mean())
```

```python
# Como mostrar media de valor de venda POR COMBUSTIVEL no Brasil todo?
media_por_combustivel_df = ca_df[['Produto', 'Valor de Venda']].groupby(by='Produto').mean().round(2)
display(media_por_combustivel_df)
```

```python
# Quero adicionar uma coluna de valor booleano no combustiveis_df
# chamada "Ativo" que, por padrão, vai ser True para TODAS as linhas.
combustiveis_df['Ativo'] = True
print(combustiveis_df.info())
display(combustiveis_df.head())
```
